#### Reduce the number of categories in the column GENDER.The column should only have either "M" for males, "F" for females, and "other" for all the rest

#### Note that there are a few null values in the column. We will first replace those null values using the code below:
#### print(categorical['GENDER'].value_counts())
#### categorical['GENDER'] = categorical['GENDER'].fillna('F')

```python
# counting the number of every value in the column 'Gender'
print(categorical['GENDER'].value_counts())
```

```python
# replacing the NaN values in the column 'Gender' with 'F'
categorical['GENDER'] = categorical['GENDER'].fillna('F')
# there is no application as we don't habe NaN values in the column Gender
```

```python
# counting again the number of every value in the column 'Gender' after replacing the NaN values in the column 'Gender' with 'F'
categorical['GENDER'].value_counts()
```

```python
# defining a function to reduce the number of categories in the column GENDER, we will have "M" for males, "F" for females
# for gaps->'' again 'F' and "other" for all the rest
def replacing_gender(x):
    if x in [' ']:
        return 'F'
    if x in ['U','J','C','A']:
        return 'other'
    else:
        return x
```

```python
# applying the function to the column Gender
# counting again the number of every value in the column 'Gender' after applying the function  
categorical['GENDER'] = categorical['GENDER'].apply(replacing_gender)
categorical['GENDER'].value_counts()
```
